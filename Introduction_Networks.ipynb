# Introduction

NetworkX is a powerful Python library designed for the creation, manipulation, and study of complex networks represented as graphs. Here are some key features and uses of NetworkX:

1. **Graph Creation and Manipulation**: You can create various types of graphs, including undirected, directed, and multigraphs. Nodes and edges can have attributes, allowing for rich data representation.
2. **Graph Algorithms**: NetworkX includes many standard graph algorithms, such as shortest path, clustering, and network flow algorithms.
3. **Network Analysis**: It provides tools to analyze the structure and dynamics of networks, including measures like centrality, connectivity, and clustering coefficients.
4. **Graph Generators**: The library includes functions to generate classic graphs (e.g., complete graphs, star graphs), random graphs, and synthetic networks.
5. **Data Structures**: NetworkX supports various data structures for graphs, making it flexible and powerful for different types of network analysis.
6. **Integration**: It integrates well with other Python libraries, such as NumPy and SciPy, for numerical computations and Matplotlib for visualization¹²³.

```python
# Example
import networkx as nx

# Create an empty graph
G = nx.Graph()

# Add nodes
G.add_node(1)
G.add_nodes_from([2, 3])

# Add edges
G.add_edge(1, 2)
G.add_edges_from([(2, 3), (3, 1)])

# Display the nodes and edges
print("Nodes:", G.nodes())
print("Edges:", G.edges())

nx.draw(G)
```

## Simple networks

It's possible define ciruclar or linear networks. A circular networks corresponds a set of nodes equisidan to a point and with only one edge.



```python
G_circular = nx.Graph()

N = 5

for i in range(N):
  G_circular.add_node(i)
lista_nodos = list(G_circular.nodes)

for i in range(N):
  G_circular.add_edge(lista_nodos[i], lista_nodos[i - 1])
nx.draw(G_circular)
```

We can also define a ring lattice to seconds, thrids and n-neighbors

```python
G_anillo = nx.Graph()

N = 25

for i in range(N):
  G_anillo.add_node(i)

lista_nodos = list(G_anillo.nodes)

for i in range(N):
  G_anillo.add_edge(lista_nodos[i], lista_nodos[i-1]) #primeros vecinos
  G_anillo.add_edge(lista_nodos[i], lista_nodos[i-2]) #segundos vecinos
  G_anillo.add_edge(lista_nodos[i], lista_nodos[i-3]) #terceros vecinos

nx.draw(G_anillo)
```

And a **Full network!**


```python
G_completa = nx.Graph()

N = 15

for i in range(N):
  for j in range(i):
    G_completa.add_edge(i,j)

nx.draw_circular(G_completa)
```
